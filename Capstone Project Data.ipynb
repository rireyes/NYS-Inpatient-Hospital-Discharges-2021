# What age group was the most common? And what was the average charge & cost?

```python
df.groupby(['Age_Group']).aggregate({'Total_Charges':'mean','Total_Costs':'mean'}).round(2)
```

# Were the discharges covered by insurance and how much? Total charges-total cost?
